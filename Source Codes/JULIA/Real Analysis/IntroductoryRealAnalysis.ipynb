#### Theorem: Cauchy Condensation Test

Let $\{a_{n}\}$ be a sequence with positive terms and decreasing.

$a_{0} \ge a_{1} \ge a_{2} \ge \dots \ge a_{n} \ge 0$

Then $\sum_{n} a_{n}$ converges if and only if

$\sum_{n} 2^{n} a_{2^{n}}$ converges.

##### Proof

Let $\{s_{n}\}$ be the sequence of partial sums of $\sum_{n} a_{n}$ and $\{S_{n}\}$ be the sequence of partial sums of $\sum_{n} 2^{n} a_{2^{n}}$

\begin{equation}
\begin{split}
a_{0} & \le a_{0} \\
a_{1} & \le a_{0} \\
a_{2} + a_{3} & \le a_{2} + a_{2} = 2a_{2} \\
a_{4} + a_{5} + a_{6} + a_{7} & \le 4a_{4} \\
\vdots \\
a_{2^{n}} + a_{2^{n}+1} + a_{2^{n}+2} + \dots + a_{2^{n+1} - 1} & \le 2^{n} a_{2^{n}}
\end{split}
\end{equation}

Then, summing $s_{2^{n+1} - 1} \le a_{0} + S_{n}$

Analogously

\begin{equation}
\begin{split}
a_{0} & \ge \frac{1}{2} a_{0} \\
a_{1} + a_{2} & \ge \frac{1}{2} 2a_{2} \\
a_{3} + a_{4} & \ge \frac{1}{2} \cdot 4 a_{4} \\
\vdots \\
a_{5} + a_{6} + a_{7} + a_{8} & \ge \frac{1}{2} 8 a_{8} \\
\vdots \\
a_{2^{n+1}+1} + a_{2^{n-1}+2} + \dots + a_{2^{n}} & \ge \frac{1}{2} 2^{n} a_{2^{n}}
\end{split}
\end{equation}

Then, summing $s_{2n} \ge \frac{1}{2} S_{n}$

$\{S_{n}\}$ is bounded $↔ \{s_{n}\}$ is bounded. The two series have the same behavior.

### Absolutely Convergent Series

#### Definition

##### Proof

$∀ p, q \ge 0$

$|a_{p} + a_{p+1} + \dots + a_{p+q}| \le |a_{p}| + |a_{p+1}| + \dots + |a_{p+1}|$

This is true also for the left-hand side:

$|a_{p} + a_{p+1} + \dots + a_{p+q}| < \varepsilon$

The Cauchy test implies the convergence of $\sum_{n} a_{n}$.

### Alternating Series

An alternating series is one in which successive terms of the sequence have opposite signs.

#### Theorem: Leibniz

Let

\begin{equation}
\sum_{n=0}^{+ \infty} (-1)^{n} a_{n} \quad \text{with} \ a_{n} > 0 \ ∀n \in \mathbb{N}
\end{equation}

If

1) the sequence $\{a_{n}\}$ is decreasing

2) $\lim_{n} a_{n} = 0$

The series $\sum_{n=0}^{+ \infty} (-1)^{n} a_{n}$ is convergent.

##### Proof

From 1)

\begin{equation}
\begin{split}
s_{n} & = a_{0} - a_{1} + a_{2} - \dots + \dots \ (-1)^{n}a_{n} \\
s_{2n+2} & = s_{2n} - (a_{2n+1} + a_{2n+2} \le s_{2n} \\
(\text{because} \ a_{2n+1} \ge a_{2n+2}) \\
s_{2n+1} & = s_{2n-1} + (a_{2n} - a_{2n+1}) \ge s_{2n-1}
\end{split}
\end{equation}

The sequence of partial sums with even indexes is decreasing.

The sequence of partial sums with odd indexes is increasing.

Furthermore,

\begin{equation}
s_{2n} - s_{2n+1} = a_{2n+1} \quad (*)
\end{equation}

The sequence $\{s_{2n}\}$ is bounded from below $\rightarrow$ it converges.

$S$ is the limit 

$S = \inf_{n} \{s_{2n}\}$

From 2)

$\lim_{n} a_{n} = 0$ and from (*) we deduce that $\{s_{2n+1}\}$ converges to $S$.

Then the series $\sum_{n=0}^{+ \infty} (-1)^{n} a_{n}$ converges and $S$ is the sum.

### Limits of Functions from $\mathbb{R}$ to $\mathbb{R}$

Let $X \subseteq \mathbb{R}$ and $f: X \rightarrow \mathbb{R}$. The aim of the limit notion is to describe the behavior of the function $f$ " close to" an accumulation point of $X$.

#### Definition

Let $f: X \rightarrow \mathbb{R}$ and $x_{0} \in \mathbb{R}^{*} = \mathbb{R} \cup \{\pm \infty \}$ an accumulation point of $X$, let $l \in \mathbb{R}^{*}$, we say that the limit of $f(x)$, for $x$ that tends to $x_{0}$ is $l$, and we write

$\lim_{x \rightarrow x_{0}} f(x) = l$

if $∀ V$ neighborhood of $l$, it is possible to find a neighborhood $U$ of $x_{0}$ such that $f(x) \in V$ if $x_{0} \neq x \in U \cap X$

### Monotone Functions

The aim is to find sufficient conditions in order to guarantee the possibility to perform the limit operation. In this sense, the calss of monotone functions is 'special' because this operation is always possible for these types of functions.

#### Theorem

i) Let $f: (a,b) \rightarrow \mathbb{R}$ be an increasing function

i1) $∀x_{0} \in (a,b]: \lim_{x \rightarrow x_{0}^{-}} f(x) = \sup_{(a,x_{0})} f$

i2) $∀x_{0} \in [a,b): \lim_{x \rightarrow x_{0}^{+}} f(x) = \inf_{(x_{0},b)} f$

ii) Let $f:(a,b) \rightarrow \mathbb{R}$ be a decreasing function

ii1) $∀x_{0} \in (a,b]: \lim_{x \rightarrow x_{0}^{-}} f(x) = \inf_{(a,x_{0})} f$

ii1) $∀x_{0} \in [a,b): \lim_{x \rightarrow x_{0}^{+}} f(x) = \sup_{(x_{0},b)} f$

### Infinitesimal and Infinite Functions

Let $f$ and $g$ be two functions defined in a neighborhood of $x_{0} \in \mathbb{R}^{*}$.

#### Definition

We say that the function $f(x)$ is infinitesimal in $x_{0}$ if $\lim_{x \rightarrow x_{0}} f(x) = 0$

We say that the function $f(x)$ is infinite in $x_{0}$ if

\begin{equation}
\lim_{x \rightarrow x_{0}} \ f(x) = + \infty \ (\text{or} - \infty) \ (\infty)
\end{equation}

#### Example

$\sin \ x, \ x^{\frac{1}{3}}, \ \log_{2} (1+x)$, these functions are infinitesimal for $x \rightarrow 0$.

* $\frac{1}{\sqrt{x}}, \ 2^{-x}$ are infinitesimal for $x \rightarrow + \infty$

If $g \neq 0$ definitely for $x \rightarrow x_{0}$, we have 4 possibilities

\begin{equation} \lim_{x \rightarrow x_{0}} \ \frac{f(x)}{g(x)}
\begin{cases}
1. 0 \\
2. l \ \text{finite number} \ l \neq 0 \\
3. (\infty, - infty)(\infty) \\
4. \nexists
\end{cases}
\end{equation}

4. holds: $f$ and $g$ are not comparable. $\lim_{x \rightarrow + \infty} \ \sin \ x \frac{2^{-x}}{2^{-x}} \nexists$

\begin{equation} \lim_{x \rightarrow x_{0}} \ \frac{f(x)}{g(x)}
\begin{cases}
1. 0 \\
2. l \ l \neq 0 \ \text{finite} \\
3. (\infty, - infty)(\infty) \\
4. \nexists
\end{cases}
\end{equation}

If

1. holds: $f$ is an infinite function with lower order with respect to $g$

2. holds: $f$ and $g$ have the same order of infinity

3. holds: $f$ is an infinite function with greater order with respect to $g$

4. holds: $f$ and $g$ are not comparable

### Landau' Symbols

1) If $g \neq 0$ definitely for $x \rightarrow x_{0}$, then

\begin{equation}
f(x) = o (g(x)) \quad \text{if} \quad \lim_{x \rightarrow x_{0}} \frac{f(x)}{g(x)} = o
\end{equation}

$o$ is a letter number 15 in the 'abc' alphabet. It is small $o$ for the equation above.

for example, $f$ is an infinitesimal function of greater order with respect to $g$ for $x \rightarrow x_{0}$.

Moreover, $f(x) = o(1)$ means that $f$ is an infinitesimal function for $x \rightarrow x_{0}$

\begin{equation}
\lim_{x \rightarrow x_{0}} \frac{f(x)}{1} = o
\end{equation}

When $\lim_{x \rightarrow x_{0}} \ f(x) = l$, if $l \in \mathbb{R}$, it implies $f(x) = l + o(1) \ x \rightarrow x_{0}$.

2) The symbol $\mathcal{O}(.)$ ("big $O$")

If $g(x) \neq 0$ definitely for $x \rightarrow x_{0}$, then

\begin{equation}
f(x) = \mathcal{O}(g(x))
\end{equation}

means that $\frac{f(x)}{g(x)}$ is definitely bounded for $x \rightarrow x_{0}$

3) The symbol $~$ (asymptote to $\dots$)

if $g(x) \neq 0$ definitely for $x \rightarrow x_{0}$, then

$f(x) ~ g(x)$ means $\lim_{x \rightarrow x_{0}} \frac{f(x)}{g(x)} = 1$

#### Remark

If $f ~ f_{1}$ and $g ~ g_{1}$ for $x \rightarrow x_{0}$ then

\begin{equation}
\frac{f}{g} ~ \frac{f_{1}}{g_{1}}
\end{equation}

this could be useful when we compute limits.

4) The symbol $\asymp$ (the function "has the same order of $\dots$")

If $g(x) \neq 0$ definitely $x \rightarrow x_{0}$, then $f(x) \asymp g(x)$ means that $∃ m$ and $M$, $0 < m < M$, such that for $x \rightarrow x_{0}$ definitely.

$m |g(x)| \le |f(x)| \le M|g(x)|$

In particular, $f(x) \asymp 1$ means that $f(x)$ and $\frac{1}{f(x)}$ are bounded in a neighborhood of $x_{0}$.

### Asymptotes

It is interesting in the following case:

\begin{equation}
f(x) = ax + b + o(1) \quad x \rightarrow + \infty \quad (\text{or} \ - \infty) \quad (*)
\end{equation}

This means that $f$ behaves like a first order polynomial when $x \rightarrow + \infty \ (- \infty)$.

In this case $y = ax+b$ is called asymptote for the function $f$.

The geometric interpretation is the following:

$\pi : \ y = ax+b$ and $P_{x}$ a point with coordinates $(x,f(x))$.

If $d(P_{x}, \pi)$ indicates the distance between $P_{x}$ and $\pi$ one has:

\begin{equation}
d(P_{x}, \pi) = \frac{(ax + b - f(x))}{\sqrt{a^{2} + 1}}
\end{equation}

$(*)$ is equivalent to say that $d(P_{x}, \pi) \rightarrow 0$ for $x \rightarrow + \infty \ (- \infty)$.

It could be easy to check:

$(*)↔ \frac{f(x)}{x} \rightarrow a \in \mathbb{R}$ and $f(x) - ax \rightarrow b \in \mathbb{R}$

#### Theorem 2: On the Zeros of a Continuous Functions

Let $f: [a,b] \rightarrow \mathbb{R}$ continuous with $f(a) \cdot f(b) < 0$, then there exists at least one $\bar{x} \in (a,b)$ such that $f(\bar{x}) = 0$

##### Proof

Consider the case $f(a) > 0$ and $f(b) < 0$

$c_{0} = \frac{a+b}{2}$ if $f(c_{0}) = 0$ you are done.

If $f(c_{0}) \neq 0$ then:

1) In the case of $f(c_{0}) < 0$

take 

$[a_{1}, b_{1}] = [a, c_{0}]$

\begin{equation}
c_{1} = \frac{a + c_{0}}{2} = \frac{a_{1} + b_{1}}{2} 
\end{equation}

if $f(c_{1}) < 0$ take

$[a_{2},b_{2}] = [a, c_{1}]$

\begin{equation}
c_{2} = \frac{a_{2} + b_{2}}{2} = \frac{a + c_{1}}{2} 
\end{equation}

if $f(c_{2}) < 0$ take

$[a_{3}, b_{3}] = [c_{2}, b_{2}]$

till $f(c_{n}) = 0$ or you have a sequence of intervals.

$[a,b] > [a_{1}, b_{1}] > [a_{2}, b_{2}] > \dots > [a_{n}, b_{n}]$

\begin{equation}
|b_{n} - a_{n}| = \frac{|b - a|}{2^{n+1}} \rightarrow 0 \quad (***)
\end{equation}

as $n \rightarrow \infty$

$a \le a_{1} \le a_{2} \le \dots \le a_{n} \le b_{n} \le b_{n-1} \le \dots \le b$

So we have two bounded monotonic sequences $\{a_{n}\}$ increasing and $\{b_{n}\}$ decreasing.

\begin{equation}
\begin{split}
∃ \alpha \in (a,b): \quad a_{n} \rightarrow \alpha \quad (*)
∃ \beta \in (a,b): \quad b_{n} \rightarrow \beta \quad (**)
\end{split}
\end{equation}

we have to prove that $\alpha = \beta$ and that this common value coincides with $\bar{x}$ which is the zero of the function $f(x)$, indeed

\begin{equation}
|\alpha - \beta| = |\alpha - a_{n} + a_{n} - b_{n} + b_{n} - \beta| \le |\alpha - a_{n}| + |a_{n} - b_{n}| + |b_{n} - \beta|
\end{equation}

* $|\alpha - a_{n}| \rightarrow 0$

* $|a_{n} - b_{n}| \rightarrow 0$

* $|b_{n} - \beta| \rightarrow 0$

thus $\alpha = \beta$

#### Theorem 3: Intermediate Values

Let $f: I \rightarrow \mathbb{R}$, $I$ interval ($(a,b)$, or $(a,b]$, or $[a,b)$, or $[a,b]$).

$f$ continuous in $I$, then $f$ takes all the values between $l \inf_{I} f(x)$ and $L = \sup_{I} f(x)$

##### Proof

Take any $t$ such that $l < t < L$, we want to prove that $∃ \bar{x} \in I$ such that $f(\bar{x}) = t$.

Since $l = \inf_{I} f(x)$ and $t>l$ by the 2nd property of $\inf$:

$∃x_{1} \in I: f(x_{1}) < t$

If $l = \inf_{I} \ f$ then

1) $∀x \in I \quad f(x) \ge l$

2) $∀ \varepsilon > 0 \ ∃ \bar{x} \in I: f(\bar{x}) < l + \varepsilon$

since $L = \sup_{I} \ f(x)$ and $t < L$ by the 2nd property of $\sup$

$∃ x_{2} \in I: f(x_{2}) > t$

thus we have

$l \le f(x_{1}) < t < f(x_{2}) \le L$

now define

$\psi: [x_{1}, x_{2}] \rightarrow \mathbb{R}$ continuous

\begin{split}

## Julia Codes to Find Tangent Line and Secant Line

* Single variable calculus has two main concepts: the derivative and the integral.

The notation is 

\begin{equation}
f'(c) = \lim_{h \rightarrow 0} \frac{f(c+h) - f(c)}{h}
\end{equation}

it is a derivative at a point.

The tangent line is most easily expressed in terms of the point-slope formula for a line where the point is $(c, f(c))$ and the slope is $f'(c)$

\begin{equation}
y = f(c) + f'(c) \cdot (x-c)
\end{equation}

## Differentiability implies Continuity

If a function is differentiable then it is continuous, not the other way around, if a function is continuous it does not guarantee that it is differentiable.

For example: function $y = |x|$ is continuous but not differentiable in $x_{0} = 0$

#### Proposition

If $f$ is differentiable in $x_{0} \in D_{f}$ then $f$ is continuous in $x_{0}$.

##### Proof

Assume $f$ is differentiable in $x_{0}$ then there exist a finite 

\begin{equation}
\lim_{x \rightarrow x_{0}} \frac{f(x) - f(x_{0})}{x - x_{0}} = f'(x_{0})
\end{equation}

Then we want to prove that

\begin{equation}
\lim_{x \rightarrow x_{0}} f(x) = f(x_{0}) ↔ \lim_{x \rightarrow x_{0}} [f(x) - f(x_{0})] = 0
\end{equation}

\begin{equation}
\begin{split}
\lim_{x \rightarrow x_{0}} [f(x) - f(x_{0})] & = \lim_{x \rightarrow x_{0}} \left( \frac{f(x) - f(x_{0})}{x - x_{0}} \right) (x - x_{0}) \\
\ & = \lim_{x \rightarrow x_{0}} \left( \frac{f(x) - f(x_{0})}{x - x_{0}}  \right) \cdot \lim_{x \rightarrow x_{0}} (x - x_{0}) \\
\ & = f'(x_{0}) \cdot 0 = 0
\end{split}
\end{equation}

with $f'(x_{0})$ is finite.

#### Example

Let $c \subset \mathbb{R}$ be a constant then $D(c) = 0$

$f(x) = c: \mathbb{R} \rightarrow \mathbb{R}$

$∀ x_{0} \in \mathbb{R}$ we will have

\begin{equation}
\lim_{x \rightarrow x_{0}} \frac{f(x) - f(x_{0})}{x - x_{0}} = \lim_{x \rightarrow x_{0}} \frac{c-c}{x - x_{0}} = \lim_{x \rightarrow x_{0}} 0 = 0
\end{equation}

#### Theorem: Basic Algebraic Properties of Derivatives

Let $f,g: I \rightarrow \mathbb{R}$ two differentiable funcitons in $x \dot{I}$ and $K \in \mathbb{R}$ be a constant.

1) $(f \pm g)' = (f)' \pm (g)'$

## Fundamental Theorems

#### Theorem: Fermat

Let $f: (a,b) \rightarrow \mathbb{R}$ and $x_{0} \in (a,b)$. If there exists the derivative of $f$ in $x_{0}$ and $f$ has a local extreme in $x_{0}$ then $f'(x_{0}) = 0$.

##### Proof

Let $x_{0}$ be a local maximum point. There exists $I$ of $x_{0}$, such that $I \subset (a,b)$, and $f(x_{0}) \ge f(x) \ ∀x \in I$.

Let us consider the

\begin{equation}
\Phi (x) = \frac{f(x) - f(x_{0})}{x - x_{0}}, \quad x \neq x_{0}
\end{equation}

For $x \in I$, we have:

(1) $x > x_{0} \rightarrow \Phi (x) \le 0$

(2) $x < x_{0} \rightarrow \Phi (x) \ge 0$

For (1),

\begin{equation}
\lim_{x \rightarrow x_{0}^{+}} \Phi (x) = f_{+}' (x_{0}) \le 0
\end{equation}

For (2),

\begin{equation}
\lim_{x \rightarrow x_{0}^{-}} \Phi (x) = f_{-}' (x_{0}) \ge 0
\end{equation}

The derivative in $x_{0}$ exists, we can conclude $f_{+}' (x_{0}) = f_{-}' (x_{0}) = 0$.

Analogously when $x_{0}$ is a local minimum point.

#### Definition

The points in which the function $f$ has a null / zero derivative are called critical points.

#### Theorem: Rolle

Let $f: [a,b] \rightarrow \mathbb{R}$, such that:

1) $f$ is continuous on $[a,b]$

2) $f$ is differentiable in $(a,b)$

3) $f(a) = f(b)$

Then there exists $c \in (a,b)$ such that $f'(c) = 0$

##### Proof
From the Weierstrass theorem we know that $f$ has a maximum and a minimum; let $x_{0}, x_{1} \in [a,b]$ such that

\begin{equation}
f(x_{0}) = M = \max_{[a,b]} \ f \quad f(x_{1}) = m = \min_{[a,b]} \ f
\end{equation}

* If $M=m$ then $f$ is a constant and $f'(x) = 0 \ ∀x \in (a,b)$

* If $M>m$ from 3) at least one between $x_{0}$ and $x_{1}$ is an interval point. In that point the derivative exists and by 2) the derivative must be zero because of the Fermat theorem.

#### Theorem: Cauchy

Let $f,g: [a,b] \rightarrow \mathbb{R}$, such that

1) $f,g$ continuous on $[a,b]$

2) $f,g$ differentiable on $(a,b)$

then there exists $c \in (a,b)$ such that

\begin{equation}
(f(b) - f(a))g'(c) = (g(b) - g(a))f'(c)
\end{equation}

##### Proof

Let us define  $w(x) = (f(b) - f(a))g(x) - (g(b) - g(a))f(x)$ then $w$ is continuous on $[a,b]$, it is differentiable on $(a,b)$ and $w(a)=w(b)$.

We can apply Rolle theorem, there exists $c \in (a,b)$ such that $w'(c) = (f(b) - f(a))g'(x) - (g(b) - g(a))f'(x) = 0$

#### Theorem: de L'Hopital

Let $ - \infty \le a < b \le + \infty$ and $f,g: (a,b) \rightarrow \mathbb{R}$ such that

a) $\lim_{x \rightarrow a^{+}} f(x) = \lim_{x \rightarrow a^{+}} g(x) = 0$ (or $+ \infty$ or $- \infty$)

b) $f, g$ are differentiable in $(a,b)$ with $g(x) \neq 0$ in $(a,b)$ with $g(x) \neq 0$ in $(a,b)$

c) $∃ \ \lim_{x \rightarrow a^{+}} \ \frac{f'(x)}{g'(x)} = L \in \bar{\mathbb{R}}$ or ($\mathbb{R}^{*}$)

then

\begin{equation}
∃ \ \lim_{x \rightarrow a^{+}} \frac{f(x)}{g(x)} = L
\end{equation}

analogous for $\lim_{x \rightarrow b^{-}}$

#### Application

#### Example

\begin{equation}
\begin{split}
\lim_{x \rightarrow 0^{+}} \left( \frac{1}{x} - \frac{\cos \ x}{\sin \ x} \right) & = \lim_{x \rightarrow 0^{+}} \frac{\sin \ x - x \ \cos \ x}{x \ \sin \ x} \quad (\text{in} \ \frac{0}{0} \ \text{form}) \\
\ & = \lim_{x \rightarrow 0^{+}} \frac{\cos \ x - \cos \ x + x \ \sin \ x}{\sin \ x + x \ \cos \ x} \cdot \frac{\sin \ x}{\sin \ x} \\
\ & = \lim_{x \rightarrow 0^{+}} \frac{x}{1 + \frac{x}{\sin \ x} \cos \ x} \\
\ & = \frac{0}{2} = 0
\end{split}
\end{equation}

#### Example

\begin{equation}
\lim_{x \rightarrow 0^{+}} x \ \ln \ x \quad \text{in} \ 0 \cdot (- \infty) \ \text{form}
\end{equation}

\begin{equation}
\begin{split}
\lim_{x \rightarrow 0^{+}} x \ \ln \ x & = \lim_{x \rightarrow 0^{+}} \frac{\ln \ x}{\frac{1}{x}} \\
\ & = \lim_{x \rightarrow 0^{+}} \frac{\frac{1}{x}}{- \frac{1}{x^{2}}} \\
\ & = \lim_{x \rightarrow 0^{+}} -x = 0
\end{split}
\end{equation}

It holds that $\lim_{x \rightarrow 0^{+}} \ x (\ln \ x)^{\alpha} = 0$ for $\alpha > 0$

## Convex Functions

#### Definition

Let $f: (a,b) \rightarrow \mathbb{R}$, it is said convex (concave) if $∀ x_{1}, x_{2} \in (a,b)$ and $∀ \alpha \in [0,1]$ it holds

1) For convex

\begin{equation}
f(\alpha x_{1} + (1 - \alpha) x_{2}) \le \alpha f(x_{1}) + (1 - \alpha) f(x_{2})
\end{equation}

2) For concave

\begin{equation}
f(\alpha x_{1} + (1 - \alpha) x_{2}) \ge \alpha f(x_{1}) + (1 - \alpha) f(x_{2})
\end{equation}

## Taylor's Formula

Let $f \in C^{n} (a,b)$ and $x_{0} \in (a,b)$, we want to approximate $f$ by a polynomial of degree $n$ in a neighborhood $U_{x_{0}}$ of $x_{0}$

This polynomial $T_{n} (f,x_{0}) \ [T^{n}(x)]$ has a contact of order $n$ with the function $f(x)$:

\begin{equation}
f^{(k)} (x_{0}) = T_{n}^{(k)} (x_{0}) \quad 0 \le k \le n \quad (*)
\end{equation}

We write $T_{n}(x)$ as

\begin{equation}
T_{n} (x) = a_{0} + a_{1} (x - x_{0}) + a_{2}(x - x_{0})^{2} + \dots + a_{n} (x - x_{0})^{n}
\end{equation}

we have to determine $a_{i}, \quad i = 0,1, \dots, n$ by $(*)$

1) $f(x_{0}) = T_{n} (x_{0}) = a_{0}$

2) $f'(x_{0}) = T_{n}'(x_{0}) = a_{1}$

\begin{equation}
T_{n}'(x) = a_{1} + 2a_{2}(x - x_{0}) + 3a_{3}(x-x_{0})^{2} + \dots + n a_{n} (x - x_{0})^{n-1}
\end{equation}

3) $f''(x_{0}) = T_{n}''(x_{0}) = 2a_{2}$

\begin{equation}
T_{n}''(x) = 2a_{2} + 3 \cdot 2 a_{3}(x - x_{0}) + \dots + n(n-1)a_{n}(x - x_{0})
\end{equation}

n) $f^{(n)}(x_{0}) = T_{n}^{(n)} (x_{0}) = n! a_{n}$

\begin{equation}
a_{n} = \frac{f^{(n)} (x_{0})}{n!}
\end{equation}

Taylor's polynomial of $f \in C^{n} (a,b)$ with $x_{0} \in (a,b)$ and $x \in U_{x_{0}}$ is

\begin{equation}
T_{n}(x, x_{0}, f) = T_{n} (x) = \sum_{k=0}^{n} \frac{f^{(k)} (x_{0})}{k!} (x - x_{0})^{k} = f(x_{0}) + f'(x_{0}) (x - x_{0}) + \frac{f''(x_{0})}{2!} (x - x_{0})^{2} + \frac{f'''(x_{0})}{3!} (x - x_{0})^{3} + \dots + \frac{f^{(n)} (x_{0})}{n!} (x - x_{0})^{n}
\end{equation}

#### Example

$f(x) = e^{x} \in C^{n} (\mathbb{R}) ∀n (\rightarrow e^{x} \in C^{\infty} (\mathbb{R}))$

$x_{0} = 0$ (MacLaurin's formula)

$f'(x) = e^{x} \quad f''(x) = e^{x} \quad \dots \quad f^{(n)}(x) = e^{x}$

\begin{equation}
T_{n} (x, e^{x}, 0) = T_{n} (x) = 1 + x + \frac{1}{2!}x^{2} + \frac{1}{3!}x^{3} + \dots + \frac{1}{n!} x^{n}
\end{equation}

```julia
# To compute multiple definite integrals

using SymPy
@vars x y z

integrate(exp(-x^2 - y^2), (x, -oo, oo), (y, -oo, oo))
```

## Important Limits of Functions

For indetermined form: $1^{\infty}$

\begin{equation}
\begin{split}
\lim_{x \rightarrow \pm \infty} \left( 1 + \frac{1}{x} \right)^{x} = e \\
\lim_{f(x) \rightarrow \pm \infty} \left( 1 + \frac{1}{f(x)} \right)^{f(x)} = e 
\end{split}
\end{equation}

For indetermined form: $\left[ \frac{0}{0} \right]$

(1)
\begin{equation}
\begin{split}
\lim_{x \rightarrow 0} \frac{a^{x} - 1}{x} = \ln \ a \\
\lim_{f(x) \rightarrow 0} \frac{a^{f(x)} - 1}{f(x)} = \ln \ a 
\end{split}
\end{equation}

(2)
\begin{equation}
\begin{split}
\lim_{x \rightarrow 0} \frac{e^{x} - 1}{x} = 1 \\
\lim_{f(x) \rightarrow 0} \frac{e^{f(x)} - 1}{f(x)} = 1 
\end{split}
\end{equation}

(3)
\begin{equation}
\begin{split}
\lim_{x \rightarrow 0} \frac{\log_{a} (1+x)}{x} = \frac{1}{\ln \ a} \\
\lim_{f(x) \rightarrow 0} \frac{\log_{a} (1+f(x))}{f(x)} = \frac{1}{\ln \ a}
\end{split}
\end{equation}

(4)
\begin{equation}
\begin{split}
\lim_{x \rightarrow 0} \frac{\ln (1+x)}{x} = 1 \\
\lim_{f(x) \rightarrow 0} \frac{\ln (1+f(x))}{f(x)} = 1
\end{split}
\end{equation}

(5)
\begin{equation}
\begin{split}
\lim_{x \rightarrow 0} \frac{(1+x)^{k} - 1}{x} = k \\
\lim_{f(x) \rightarrow 0} \frac{(1+f(x))^{k} - 1}{f(x)} = k
\end{split}
\end{equation}

(6)
\begin{equation}
\begin{split}
\lim_{x \rightarrow 0} \frac{\sin \ x}{x} = 1 \\
\lim_{f(x) \rightarrow 0} \frac{\sin(f(x))}{f(x)} = 1
\end{split}
\end{equation}

(7)
\begin{equation}
\begin{split}
\lim_{x \rightarrow 0} \frac{1 - \cos \ x}{x^{2}} = \frac{1}{2} \\
\lim_{f(x) \rightarrow 0} \frac{1 - \cos(f(x))}{[f(x)]^{2}} = \frac{1}{2}
\end{split}
\end{equation}

(8)
\begin{equation}
\begin{split}
\lim_{x \rightarrow 0} \frac{\tan \ (x)}{x} = 1 \\
\lim_{f(x) \rightarrow 0} \frac{\tan(f(x))}{f(x)} = 1
\end{split}
\end{equation}
