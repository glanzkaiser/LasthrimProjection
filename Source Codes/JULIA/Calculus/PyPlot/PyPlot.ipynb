# PyPlot 

PyPlot backend for Plots in Julia is interactive and can be rotated. It is useful for those curious enough to see from all angles 3-d plot of a surface or any kind of plots.

```julia
using Pkg

Pkg.add(url="https://github.com/matthiashimmelmann/Implicit3DPlotting.jl.git")
```

```julia
using Pkg

Pkg.add("PyPlot")
```

```julia
# Testing plot

using Plots
pyplot()

xc = [0,0,1]
yc = [0,1,0]
zc = [1,0,0]

plot(xc,yc,zc,st=:line,arrows=:true,camera=(-30,30))
```

```julia
using Plots
pyplot()

o=(0,0,0)
x=(1,0,0)
y=(0,1,0)
z=(0,0,1)

PyPlot.plot3D(o,x,y,z)
```

```julia
using Plots
pyplot()

x=range(-2,stop=2,length=100)
y=range(sqrt(2),stop=2,length=100)
f(x,y) = x*y-x-y+1
plot(x,y,f,st=:surface,camera=(-30,30))
```

```julia
using Plots
pyplot()

xc = [0,0,1,1]
yc = [0,1,1,0]
zc = [1,1,2,2]

plot(xc,yc,zc,st=:surface,camera=(-30,30))
```

```julia
using Plots
pyplot()

x=range(-2,stop=2,length=100)
y=range(sqrt(2),stop=2,length=100)
my_cg = cgrad([:green,:blue])

f(x,y) = ((x^(3) - x) + (y^(3) - y))^(0.5)

plot(x,y,f,st=:surface,c=my_cg,camera=(-30,30))
```

```julia
# if you want to use another heatmap, just set the color (or c) argument to some color gradient

using Plots
pyplot()

x=range(-2,stop=2,length=100)
y=range(sqrt(2),stop=2,length=100)
my_cg = cgrad([:green,:blue])

f(x,y) = x*y-x-y+1

plot(x,y,f,st=:surface,c=my_cg,camera=(-30,30))
```
