What is a differential equation?

$A \subset \mathbb{R^{n+2}}$ an open set

$A$ open ↔ for each $x_{0} \in A$ there is $\varepsilon > 0$ and $\beta _{\varepsilon} (x_{0}) \subset A$

Im $\mathbb{R^{2}}: \{(x,y | x \ge 0, y \ge 0)\}$ is not open.

$\{(x,y) | x > 0, y > 0\}$ is open

#### Definition
Take $A \subset \mathbb{R^{n+2}}$ and $F: A \subset \mathbb{R^{n+2}} → \mathbb{R}$ continuous function.

Then 

$F(t, y, y', y'', \dots, y^{(n)}) = 0$ 

is a differential equation of order $n$.

Here $y = y(t)$ is a funciton and $y^{(k)}$ is the $k$th-derivative of $y$

\begin{equation}
y^{(k)} (t) = \frac{d^{k}}{dt^{k}} y(t)
\end{equation}

$y$ is the unknown of the differential equation problem, while $t$ is called independent variable.

We will concentrate on the special case:

This differential equation is called in normal form("highest derivative order of $y$")

\begin{equation}
y^{(n)} = f(t, y, y', \dots, y^{(n-1)})
\end{equation}

where $f: \Omega \subset \mathbb{R^{n+1}} \rightarrow \mathbb{R}$ 

$\Omega$ open set, and $f$ is a continuous function.

The highest derivative order of $y$ = Function of $t, y, y', \dots, y^{(n-1)}$ up to an order less than the highest.

#### Example

\begin{equation}
y' = f(t) \\
y(t_{0}) = y_{0}
\end{equation}

$f$ is a continuous function.

$f: I \subseteq \mathbb{R} → \mathbb{R}$ and $t_{0} \in I$.

Integrate between $t_{0}$ and $t$ ($t > t_{0}$).

\begin{equation}
\int_{t_{0}}^{t} y' (s) ds = y(t) - y(t_{0}) \\
\int_{t_{0}}^{t} f(s) ds 
\end{equation}

\begin{equation}
y(t) = y_{0} + \int_{t_{0}}^{t} f(s) ds 
\end{equation}

This is the solution to the problem alone. One has existence and uniqueness of the solution by the Fundamental Theorem of Calculus.

#### Example 2

\begin{equation}
y' = y \\
y(0) = 2
\end{equation}

$y = y(t)$ and $t$ is the independent variable.

Claim: $y(t) = 2 e^{t}$ is a solution.

Substitute to the initial condition formula:

$y(0) = 2 e^{0} = 2$ (proven)

\begin{equation}
\frac{d}{dt} y(t) = \frac{d}{dt} (2e^{t}) = 2e^{t} \\
y' = y
\end{equation}

$m \ddot{x} = F(t, x, \dot{x})$

$x = x(t)$ is the unknown

\begin{equation}

has a unique solution from the physical point of view.

Some hypothesis is required on $F$ to guarantee that the problem ($N$) has a unique solution.

### Cauchy Problem of $n$-th order Equation

\begin{equation}
\begin{cases}
y^{(n)} = f(t, y, y', \dots, y^{(n-1)}) \\
y(t_{0}) = y_{0} \in \mathbb{R} \\
\vdots \\
y^{(n-1)} (t_{0}) = y_{n-1} \in \mathbb{R}
\end{cases}
\end{equation}

$y = y(t)$ is a function (unknown).

The problem above is called Cauchy problem of the $n$-th order.

In general, for a problem of order $n$, $n$ initial conditions are needed to hope to have a unique solution.

What are the conditions that give existence and uniqueness of the solution?

### Cauchy Problem for First Order Equations

\begin{equation} (c)
\begin{cases}
y' = f(t,y) \quad \Omega \ \text{an open set} \\
y(t_{0}) = y_{0} \quad f: \Omega \subset \mathbb{R}^{2} \rightarrow \mathbb{R} 
\end{cases}
\end{equation}

$y = y(t)$ is the solution to $(c)$ if we are in the hypothesis of existence and uniqueness and 

$y: (t_{1}, t_{2}) \rightarrow \mathbb{R}$

where $(t_{1}, t_{2}) is the maximal interval of existence$

#### Example

$\Omega = (- \frac{\pi}{2}, \frac{\pi}{2}) \times \mathbb{R}$

\begin{equation} (p)
\begin{cases}
y' = y^{2} + 1, \quad f(t,y') = y^{2} + 1 \\
y_{0} = 0 \quad ( y(t_{0}) = y_{0} ) \\
\end{cases}
\end{equation}

* Claim:

$y(t) = \tan(t)$ is a solution to $(p)$.

$\tan(t) = \frac{\sin(t)}{\cos(t)}$

$y(0) = \tan (0) = 0$ (proven)

\begin{equation}
\begin{split}
\frac{d}{dt} (y(t)) & = \frac{d}{dt} (\tan(t)) \\
\ & = \frac{d}{dt} \frac{\sin(t)}{\cos(t)} \\
\ & = \frac{d}{dt} \frac{\sin(t) \cdot \sin(t) + \cos^{2} (t)}{\cos^{2}(t)} \\
\ & = \frac{1}{\cos^{2}(t)}
\end{split}
\end{equation}

\begin{equation}
\begin{split}
1 + y^{2}(t) & = 1 + (\tan(t))^{2}\\
\ & = \frac{\cos^{2}(t) + \sin^{2}(t)}{cos^{2}(t)} \\
\ & = \frac{1}{\cos^{2}(t)}
\end{split}
\end{equation}

Then the claim is shown

### Local Existence and Uniqueness Theorems

#### Theorem 1 (Local existence, PEANO Theorem)

\begin{equation} (c)
\begin{cases}
y' = f(t,y) \\
y(t_{0}) = y_{0} 
\end{cases}
\end{equation}

$f$ continuous in $\Omega$

$f: \Omega \subset \mathbb{R_{t}} \times \mathbb{R_{y}} → \mathbb{R}$

$\Omega$ open set $(t_{0}, y_{0}) \in \Omega$

Then the Cauchy problem $(c)$ admits at least one solution $y = y(t)$ defined on its maximal interval of existence.

$I = I(t_{0}) = (t_{1}, t_{2})$, where $y: I \subseteq \mathbb{R} → \mathbb{R}$ and $y \in c^{1}(I)$

* Terminology: a function $g: I → \mathbb{R}$ is of class $c^{k} (I)$, $k$ positive integer if $g$ is continuous on $I$ and all the derivatives up to order $k$ exist and are continuous in $I$.

#### Example

\begin{equation}
f(x) =
\begin{cases}
x^{2} \sin(\frac{1}{x}) \quad x \neq 0 \\
0 \quad x = 0 
\end{cases}
\end{equation}

$f: \mathbb{R} \rightarrow \mathbb{R}$

is $f$ a function of class $c^{1} (\mathbb{R})$ ?

* You can check that $f(x)$ is continuous.

$f'(x) = 2x \sin(\frac{1}{x}) - cos(\frac{1}{x})$ if $x \neq 0$

\begin{equation}
\begin{split}
f'(0) & = \lim_{h→0} \frac{f(h) - f(0)}{h}\\
\ & = \lim_{h→0} \frac{h^{2} \sin(\frac{1}{h}) - 0}{h}\\
\end{split}
\end{equation}

\begin{equation}f'(x)
\begin{cases}
2x \ \sin(\frac{1}{x}) - \cos(\frac{1}{x}) \quad \text{if} \ x \neq 0 \\
0 \quad \text{if} x = 0
\end{cases}
\end{equation}

$f'(x)$ is not continuous.

$\cos(\frac{1}{x}) \nrightarrow 0$ when $x \rightarrow 0$.

#### Theorem 2 (Local Existence and Uniqueness, Cauchy-Lipschitz Theorem)

\begin{equation}(c)
\begin{cases}
y' = f(t,y) \\
y(t_{0}) = y_{0} 
\end{cases}
\end{equation}

$f: \Omega \subset \mathbb{R_{t}} \times \mathbb{R_{y}} \rightarrow \mathbb{R}$

$\Omega$ an open set

$(t_{0}, y_{0}) \in \Omega$

Suppose that $∃L > 0$ such that $|f(t, y_{2} - f(t, y_{1})| \le L |y_{1} - y_{2}|$ for all $(t, y_{1}), (t, y_{2}) \in \Omega$.

Assume also that $f$ is continuous in $\Omega$.

Then the Cauchy problem $(c)$ admits an unique solution $y = y(t)$ defined in the maximal interval of existence $I = I(t_{0}) =(t_{1}, t_{2})$ with $t_{1} < t_{0} < t_{2}$ and $y \in c^{1} (I)$.

This solution $y$ is unique in any sub-interval $ I \subseteq I(t_{0})$

* The condition: " $∃L > 0$ " is called Lipschitz continuity with respect to the $y$ variable.

The Lipschitz continuity condition with respect to $y$ can be written as $∃L > 0$ such that

\begin{equation}
\frac{|f(t,y_{2} - f(t, y_{1}|}{|y_{1} - y_{2}|} \le L
\end{equation}

for all $(t, y_{1} \neq (t, y_{2})$

$(t, y_{1}, (t, y_{2}) \in \Omega$

#### Counter-example for Theorem 2

\begin{equation}
\begin{cases}
y' = \sqrt{|y|} \\
y_{0} = 0 
\end{cases}
\end{equation}

$f(y) = \sqrt{|y|}$

$f: \mathbb{R} → \mathbb{R}$ continuous.

$y > 0 \quad f(y) = \sqrt{y}$ and $f'(y) = \frac{1}{2\sqrt{y}}$

In any neighborhood of $y_{0} = 0$ one has that does not exist on $L > 0$ independent of the $y$ variable such that $|\sqrt{|y|}| \le L|y|$

Hence the Lipschitz continuity condition of the Cauchy-Lipschitz theorem is not satisfied.

On the other hand, the hypothesis for the existence of a solution is satisfied (Peano theorem holds)

* Some solutions

1) $y(t) = 0$

$y(0) = 0$

$y' = 0$ and $\sqrt{|y|} = 0$

Hence this is a solution

2) 
\begin{equation}
y(t)
\begin{cases}
0 \quad t \le 0\\
\frac{1}{4} t^{2} \quad t > 0 
\end{cases}
\end{equation}

$y(0) = 0$ (proven)

\begin{equation}
y'(t)
\begin{cases}
0 \quad t \le 0\\
\frac{1}{2} t \quad t > 0 
\end{cases}
\end{equation}

$ \sqrt{\frac{1}{4}t^{2}} = \frac{1}{2} t$ for $t = 0$

Hence, the equation is satisfied also for this second solution.

### Method for Solving Nonlinear ODEs: Separation of Variables

\begin{equation}
(*)
\begin{cases}
y' =h(t) g(y) \quad \\
y(t_{0}) = y_{0}
\end{cases}
\end{equation}

$f(t,y) = h(t) g(y)$

To satisfy the Cauchy-Lipschitz assumptions let us assume that $g \in C^{1}(I)$ and $h \in C^{1} (J)$.

$I \subset \mathbb{R}, \quad J \subset \mathbb{R}$ open interval

#### The Method

If $y_{0}$ is such that $g(y_{0}) = 0$, then the unique solution to (*) is $y(t) = y_{0}$;

Otherwise, $g(y_{0}) \neq 0$. Furthermore let us call $\tilde{y}(t)$ the unique solution to (*).

Then, if $g(y_{0}) \neq 0$, then $g(\tilde{y}(t)) \neq 0$ for all $t$ in a neighborhood of $t_{0}$.

Hence 

\begin{equation}
\frac{\tilde{y'}(t)}{g(\tilde{y} (t))} = h(t)
\end{equation}

for all $t$ in a neighborhood of $t_{0}$

Integrate the equality above in $(t_{0}, t)$

\begin{equation}
\int_{t_{0}}^{t} \frac{\tilde{y}'(s)}{g(\tilde{y} (s))} ds = \int_{t_{0}}^{t} h(s) ds
\end{equation}

for all $t$ in a neighborhood of $t_{0}$

\begin{equation}
z =\tilde{y} (s) ↔ dz = \tilde{y}' (s) ds
\end{equation}

\begin{equation}
\int_{y_{0}}^{\tilde{y}(t)} \frac{dz}{g(z)} = \int_{t_{0}}^{t} h(s) ds
\end{equation}

If we can solve the integrals above, then we can obtain an analytic solution.

$ \frac{A(1-z) + B(1+z)}{(1+z)(1-z)} = \frac{A+B+Z(B-A)}{1-z^{2}} $

\begin{equation}
\begin{cases}
A + B = 1 \quad \rightarrow \quad 2A = 1 \rightarrow A = \frac{1}{2} \\
B - A = 0 \quad \rightarrow \quad B = A \rightarrow B = \frac{1}{2}
\end{cases}
\end{equation}

\begin{equation}
\begin{split}
\int_{0}^{\tilde{y} (t) \frac{dz}{1 - z^{2}}} & = \frac{1}{2} \int_{0}^{\tilde{y} (t)} \frac{dz}{1+z} + \frac{1}{2} \int_{0}^{\tilde{y} (t)} \frac{dz}{1 - z} \\
\ & = \frac{1}{2} [\ln |1+z|]_{0}^{\tilde{y} (t)} - \frac{1}{2} [\ln |1-z|]_{0}^{\tilde{y} (t)} \\

\ & = \frac{1}{2} \ln (1 + \tilde{y} (t)) - \frac{1}{2} \ln (1 - \tilde{y} (t)) \\
\ & = \frac{1}{2} \ln \left( \frac{1 + \tilde{y} (t)}{1 - \tilde{y} (t)} \right)
\end{split}
\end{equation}

## Linear Differential Equations of Order $n$

\begin{equation}
y^{n} + a_{n-1} (t) y^{(n-1)} + \dots + a_{1} (t) y' (t) + a_{0} (t) y(t) = g(t) \quad (*)
\end{equation}

where $t \in I \subseteq \mathbb{R}$ open interval.

$a_{j}, g: I \rightarrow \mathbb{R}$ continuous functions with $j=0,1,\dots,n-1$

* $a_{j}$ coefficients (non-constant if they time-dependent)

* $g$ non-homogeneous term (if $g=0$, then the problem is called homogeneous)

We are looking for $y = y(t)$ solution on $I$ and such that $y$ is of class $C^{n} (I)$. Furthermore, $y$ is always global.

### Properties Coming from the Linearity

Suppose that $y_{1} (t)$ and $y_{2} (t)$ are solutions to $(*)$:

\begin{equation}
y_{1}^{n} + a_{n-1} (t) y_{1}^{(n-1)} + \dots + a_{1} (t) y_{1}' (t) + a_{0} (t) y_{1}(t) = g(t) 
\end{equation}

and

\begin{equation}
y_{2}^{n} + a_{n-1} (t) y_{2}^{(n-1)} + \dots + a_{1} (t) y_{2}' (t) + a_{0} (t) y_{2}(t) = g(t) 
\end{equation}

call $w(t) = y_{1} (t) - y_{2} (t)$ 

it follows from linearity that

\begin{equation}
w^{(n)} (t) + a_{n-1} w^{(n-1)} + \dots + a_{1} w'(t) + a_{0} (t) w(t) = 0
\end{equation}

Hence $w(t)$ satisfies $(*)$ with $g=0$ (homogeneous version of $(*)$), that is 

\begin{equation}
y^{n} + a_{n-1} (t) y^{(n-1)} + \dots + a_{1} (t) y' (t) + a_{0} (t) y(t) = 0 \quad (**)
\end{equation}

This is a linear $n$th-order homogeneous differential equations

#### Proposition

(a) Given $y_{1}$ and $y_{2}$ solutions to $(**)$, then any linear combination is still a solution, that is

\begin{equation}
\lambda_{1} y_{1} (t) + \lambda_{2} y_{2} (t)
\end{equation}

is still a solution with $\lambda_{1}, \lambda_{2} \in \mathbb{R}$.

(b) Given $y_{1}, y_{2}$ solutions to $(*)$, then $y_{1} - y_{2}$ is a solution to $(**)$. Furthermore if $\bar{y} (t)$ is a solution to $(*)$ and $w(t)$ is a solution to $(**)$, then

\begin{equation}
w(t) + \bar{y} (t)
\end{equation}

is still a solution to $(*)$.

#### Remark

We can split the problem of looking for a solution to $(*)$ into two steps:

1) Find one solution to $(*)$

2) Find the general integral for $(**)$

## Cauchy Problem Related to (*)

\begin{equation}
\begin{cases}
y^{(n)} + a_{n-1} y^{(n-1)} + \cdots + a_{1}y' + a_{0}y = g \\
y(t_{0}) = y_{0} \quad \quad (\text{Problem}) \ 1 \\
\vdots \\
y^{(n-1)} (t_{0}) = y_{n-1}
\end{cases}
\end{equation}

$t \in I, \ t_{0} \in I, \ (y_{0}, y_{1}, \dots, y_{n-1}) \in \mathbb{R}^{n}$

$I$ is open interval.

One can show that given $t_{0} \in I, \ (y_{0}, y_{1}, \dots, y_{n-1}) \in \mathbb{R}^{n}$ then (Problem 1) has a unique solution (globally defined).

Now consider the homogeneous problem.

\begin{equation}
\begin{cases}
y^{(n)} + a_{n-1} y^{(n-1)} + \cdots + a_{1}y' + a_{0}y = 0 \\
y(t_{0}) = y_{0} \quad \quad (\text{Problem}) \ 2 \\
\vdots \\
y^{(n-1)} (t_{0}) = y_{n-1}
\end{cases}
\end{equation}

$t \in I, \ t_{0} \in I, \ (y_{0}, y_{1}, \dots, y_{n-1}) \in \mathbb{R}^{n}$

How can I find a general integral to this problem?

Consider the following homogeneous problem

\begin{equation}
\begin{cases}
y^{(n)} + a_{n-1} y^{(n-1)} + \cdots + a_{1}y' + a_{0}y = 0 \\
y(t_{0}) = y_{0} \quad \quad (\text{Problem}) \ K \\
y'(t_{0}) = 0 \\
\vdots \\
y'^{(K)} (t_{0}) = 1 \\
\vdots \\
y^{(n-1)} (t_{0}) = 0
\end{cases}
\end{equation}

$K = 0,1, \dots, n-1$

Observe that

\begin{equation}y^{(m)} (t_{0})
\begin{cases}
0 \quad m \neq K \\
1 \quad m = K
\end{cases}
\end{equation}

and this set of initial condition is a basis in $\mathbb{R}^{n}$.

Observe that (Problem K) has also a unique solution for each $K=0,1, \dots, n-1$

Hence, (Problem 2) has a unique global solution given by

\begin{equation}
y(t) = \sum_{k=0}^{n-1} y_{K} \mu_{K} (t)
\end{equation}

where $y_{K}, \ k=0, 1, \dots, n-1$ are the initial conditions of (Problem 2) and $\mu_{K}(t)$ are the solutions to (Problem K), with $K=0,1,\dots,n-1$.

Hence, $\{\mu_{K} (t)\}_{K=0,1,\dots,n-1}$ generate any solution to

\begin{equation}
y^{(n)} + a_{n-1} (t) y^{(n-1)} (t) + \dots + a_{1} (t)y'(t) + a_{0} (t) y(t) = 0
\end{equation}

Furthermore, one can show that $\{\mu_{K}\}_{K=0,1,\dots,n-1}$ are linearly independent, meaning that

\begin{equation}
\sum_{k=0}^{n-1} \alpha_{K} \mu_{K} (t) = 0 \rightarrow \alpha_{K} = 0
\end{equation}

for all $K$.

Note that:

\begin{equation}
\begin{split}
\mu_{0} (t_{0}) = 1, & \mu_{0}' (t_{0}) = 0, & \dots, & \mu_{0}^{(n-1)} (t_{0}) = 0 \\
\mu_{1} (t_{0}) = 0, & \mu_{1}' (t_{0}) = 1, & \dots, & \mu_{1}^{(n-1)} (t_{0}) = 0 \\
\vdots
\end{split}
\end{equation}

Hence to verify that $\{\mu_{K}\}$ is linearly independent, then

$0 = \sum \alpha_{K} \mu_{K} (t)$

then we can evaluate at $t=t_{0}$

\begin{equation}
\begin{split}
0 & = \sum_{K=0}^{n-1} \alpha_{K} \mu_{K} (t_{0}) = \alpha_{0} \\
\alpha_{0} & = 0
\end{split}
\end{equation}

Then compute the derivation of

$\sum_{K=0}^{n-1} \alpha_{K} \mu_{K} (t)$

\begin{equation}
\begin{split}
0 & = \sum_{K=0}^{n-1} \alpha_{K} \mu_{K}' (t_{0}) = \alpha_{1}\\
\vdots
\end{split}
\end{equation}

One keeps repeating this prochedure, obtaining $\alpha_{0}, \dots, \alpha_{n-1} = 0$

#### Proposition

The set of all solutions to the differential equation

\begin{equation}
y^{(n)} (t) + a_{n-1} (t) y^{(n-1)} (t) + \dots + a_{1} (t) y' (t) + a_{0} (t) y(t) = 0
\end{equation}

is a linear space of dimension $n$ ($n$ = order of the equation).

A basis for such a linear space is given by the solutions $\mu_{K}, \ K=0,1, \dots, n-1$ corresponding to the (Problem K) Cauchy Problem.

Moreover, the solution to (Problem 2) is given by

\begin{equation}
y(t) = \sum_{K=0}^{n-1} y_{K} \mu_{K} (t)
\end{equation}

Finally, if we have $\bar{y} (t)$ a given solution to equation $(*)$ ($n$-th order linear non homogeneous) then the general integral to $(*)$ is given by

\begin{equation}
\bar{y}(t) = \sum_{K=0}^{n-1} \alpha_{K} \mu_{K} (t)
\end{equation}

with

$\alpha_{0}, \alpha_{1}, \dots, \alpha_{n-1} \in \mathbb{R}$

## Homogeneous Linear Differential Equations of Order $n$ and Constant Coefficients

\begin{equation}
y^{(n)} (t) + a_{n-1} y^{(n-1)} (t) + \dots + a_{1}y' (t) + a_{0} y(t) = 0 \quad (HC)
\end{equation}

Suppose we look for a solution to this problem in this given form

\begin{equation}
y(t) = e^{\lambda t} \ , \quad \lambda \in (\mathbb{C})
\end{equation}

\begin{equation}
\begin{split}
y' (t) & = \lambda e^{\lambda t} \\
y^{(K)} (t) & = \lambda^{K} e^{\lambda t} \\
\vdots \\
y^{(n)} (t) & = \lambda^{n} e^{\lambda t} \\
\end{split}
\end{equation}

Replace this identities in $(HC)$ and observe that $e^{\lambda t} \neq 0$ for all $\lambda, t \in \mathbb{R}$

\begin{equation}
\lambda^{n} + a_{n-1} \lambda^{n-1} + \dots + a_{1} \lambda + a_{0} = 0 \quad (**)
\end{equation}

This argument shows that $y(t) = e^{\lambda t}$ is a solution if and only if $\lambda$ is a root of $(**)$.

It follows from the Fundamental Theorem of Algebra that $(**)$ has $n$ solutions in $\mathbb{C}$

1) Case 1:

We have $n$ solutions to $(**)$ real and distinct.

As a consequence

$e^{\lambda_{0} t}, e^{\lambda_{1} t}, \dots, e^{\lambda_{n-1} t}$ are all solutions to $(HC)$

and

\begin{equation}
y(t) = \sum_{i=0}^{n-1} c_{i} e^{\lambda_{i} t}
\end{equation}

the equation above is the general integral with $c_{0}, c_{1}, \dots, c_{n-1} \in \mathbb{R}$

(Notice that $e^{\lambda_{0} t}, \dots, e^{\lambda_{n-1} t}$ are linearly independent functions)

2) Case 2:

Suppose that we have $K$ real distinct solutions $\lambda_{1}, \lambda_{2}, \dots, \lambda_{K}$ and suppose that their multiplicity is $\eta_{1}, \eta_{2}, \dots, \eta_{K}$ with $\sum_{i=1}^{K} \eta_{i} = n$.

In this case, one can show that
\begin{equation}
\begin{split}
e^{\lambda_{1} t}, & t e^{\lambda_{1} t}, & \dots, & t^{(\eta_{1} - 1)} e^{\lambda_{1}t} \\
e^{\lambda_{2} t}, & t e^{\lambda_{2} t}, & \dots, & t^{(\eta_{2} - 1)} e^{\lambda_{2}t} \\
\vdots \\
e^{\lambda_{K} t}, & t e^{\lambda_{K} t}, & \dots, & t^{(\eta_{K} - 1)} e^{\lambda_{K}t} 
\end{split}
\end{equation}

are all solutions to $(HC)$ and the are linearly independent. Furthermore, since they are $n$, then they are a basis in $\mathbb{R}^{n}$.

The general integral in this case is

\begin{equation}
y(t) = \sum_{\alpha=1}^{K} \left( \sum_{\beta = 0}^{\eta_{\alpha} - 1} C_{\alpha, \beta} t^{\beta} e^{\gamma_{\alpha} t} \right)
\end{equation}

$C_{\alpha, \beta} \in \mathbb{R}$ (there are $n$ constants)

$\alpha = 1, 2, \dots, K, \ \beta = 0, 1, \dots, \eta_{\alpha} - 1$

3) Case 3:

Suppose that among the solutions to $(**)$, we have a couple of complex and conjugate solutions

\begin{equation}
\lambda = \alpha \pm i \beta
\end{equation}

In this case we can verify that

\begin{equation}
e^{\alpha t} \ \cos(\beta \ t), e^{\alpha t} \ \sin(\beta \ t) 
\end{equation}

are solutions to $(HC)$ and they are linearly independent. Furthermore if $\lambda = \alpha \pm i \beta$ has multiplicity $\nu$, then we have as solutions to $(HC)$ the following equations:

\begin{equation}
\begin{split}
e^{\alpha t} \ \cos(\beta t), & e^{\alpha t} \sin(\beta t) \\
te^{\alpha t} \ \cos(\beta t), & te^{\alpha t} \sin(\beta t) \\
\vdots
t^{\nu - 1}e^{\alpha t} \ \cos(\beta t), & t^{\nu - 1}e^{\alpha t} \sin(\beta t) \\
\end{split}
\end{equation}

thus, we have $2 \nu$ linearly independent solutions.

the equation above is homogeneneous. With $y=y(t) \in C^{n}(\mathbb{R})$ it is homogeneous linear differential equation with constant coefficients, $n$-th order

$a_{0}, a_{1}, \dots, a_{n-1} \in \mathbb{R} ↔$ constant coefficients.

## Characteristic Equation Associated to $(*)$

\begin{equation}
\lambda^{(n)} + a_{n-1} \lambda^{(n-1)} + \dots + a_{1}\lambda + a_{0}y = 0 \quad (**)
\end{equation}

$\lambda \in \mathbb{R} (\mathbb{C})$

Summarizing the results that we discussed in course 5, one has the following:

#### Theorem: General Integral for the Homogeneous Linear Equation of Order $n$

Consider the equation $(*)$ and the corresponding characteristic equation $(**)*. Suppose that $(**)$ admits:

i) $k$ distinct and real solutions

$\lambda_{1} < \lambda_{2} < \lambda_{3} < \dots < \lambda_{k}$

with multiplicity

$\nu_{1}, \nu_{2}, \nu_{3}, \dots, \nu_{k} \in \mathbb{N}$

respectively;

ii) The complex and conjugated solutions to $(**)$:

$\alpha_{1} \pm i \beta_{1}, \dots, \alpha_{n} \pm i \beta_{n}$

with multiplicity

$s_{1}, s_{2}, \dots, s_{n} \in \mathbb{N}$

respectively (we have the same multiplicity for the solutions with + and -).

Then a basis for the linear space of the solutions to $(*)$ is given by:

\begin{equation}
\begin{split}
e^{\lambda_{1} t}, & te^{\lambda_{1} t}, \dots, t^{\nu_{1}-1}e^{\lambda_{1} t} \\
e^{\lambda_{2} t}, & te^{\lambda_{2} t}, \dots, t^{\nu_{2}-1}e^{\lambda_{2} t} \\
\vdots \\
e^{\lambda_{k} t}, & te^{\lambda_{k} t}, \dots, t^{\nu_{k}-1}e^{\lambda_{k} t} \\
\end{split}
\end{equation}

\begin{equation}
\begin{split}
e^{\alpha_{1} t} \cos(\beta_{1} t), & e^{\alpha_{1} t} \sin(\beta_{1} t)\\
te^{\alpha_{1} t} \cos(\beta_{1} t), & te^{\alpha_{1} t} \sin(\beta_{1} t)\\
\vdots \\
(t)^{s_{1}-1} e^{\alpha_{1}} \cos(\beta_{1} t), & t^{(s_{1}-1)}e^{\alpha_{1}} \sin(\beta_{1} t)\\
e^{\alpha_{2} t} \cos(\beta_{2} t), & e^{\alpha_{2} t} \sin(\beta_{2} t) \\
te^{\alpha_{2} t} \cos(\beta_{2} t), & te^{\alpha_{2} t} \sin(\beta_{2} t) \\
\vdots \\
(t)^{s_{2}-1} \cos(\beta_{2} t), & (t)^{s_{2} - 1} \sin(\beta_{2} t) \\
\vdots \\
e^{\alpha_{n} t} \cos(\beta_{n} t), & e^{\alpha_{n} t} \sin(\beta_{n} t) \\
te^{\alpha_{n} t} \cos(\beta_{n} t), & te^{\alpha_{n} t} \sin(\beta_{n} t) \\
\vdots \\
t^{s_{n} -1} e^{\alpha_{n} t} \cos(\beta_{n} (t)), & t^{s_{n} - 1} e^{\alpha_{n} t} \sin(\beta_{n} t)
\end{split}
\end{equation}

## Lagrange Method (Variation of Constants)

\begin{equation}
y^{(n)} + a_{n-1} y^{(n-1)} + \dots + a_{1} y' + a_{0} y = g \quad (1)
\end{equation}

$I$ open interval

$a_{n}, g \in C(I)$

We know that the homogeneous equation

\begin{equation}
y^{(n)} + a_{n-1} y^{(n-1)} + \dots + a_{1}y' + a_{0}y = 0 \quad (2)
\end{equation}

admits $n$ linearly independent solutions (basis).

\begin{equation}
\{y_{1} (t), \dots, y_{n} (t)\}
\end{equation}

The general integral to $(2)$ is given by

\begin{equation}
y(t) = \sum_{\alpha=1}^{n} k_{\alpha} y_{\alpha} (t)
\end{equation}

$k_{\alpha} \in \mathbb{R}, \ \alpha = 1, 2, \dots, n$

We now look for a solution to $(1)$ in this form

\begin{equation}
y(t) = \sum_{\alpha=1}^{n} k_{\alpha} (t) y_{\alpha} (t) \quad (3)
\end{equation}

The new unknown are the $k_{\alpha}$ functions, $\alpha = 1, \dots, n$.

Let us impose $y(t)$ solution to $(1)$ and compute the derivatives of $y(t)$.

Derivate $(3)$

\begin{equation}
\begin{split}
y'(t) & = \sum_{\alpha = 1}^{n} (k_{\alpha} (t) y_{\alpha} (t))' \\
\ & = \sum_{\alpha = 1}^{n} k_{\alpha}' (t) y_{\alpha} (t) + \sum_{\alpha=1}^{n} k_{\alpha} (t) y_{\alpha}' (t) \\
y''(t) & = \sum_{\alpha = 1}^{n} k_{\alpha}' (t) y_{\alpha}' (t) + \sum_{\alpha=1}^{n} k_{\alpha} (t) y_{\alpha}''(t)
\end{split}
\end{equation}
